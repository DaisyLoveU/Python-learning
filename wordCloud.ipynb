```
# 内容分词
import jieba
lyst = list(text)
content_s = jieba.lcut(text)
content = [i for i in content_s if len(i)>1 and i != '\r\n']  
```

```
# 去停用词
content_clean = []
all_words = []
stop_words = list(cut)
for word in content:
    if word in stop_words:
        continue
    content_clean.append(word)
    all_words.append(str(word))
# 统计词频
words_count = {}
for word in content_clean:
    if word not in words_count:
        words_count[word] = 1
    else:
        words_count[word] += 1

word_frequence = dict(sorted(words_count.items(), key=lambda x:x[1], reverse=True))
```

```
# 生成词云  需要下载中文字体，放在相应目录
word_cloud = wordcloud.WordCloud(background_color='white', max_font_size=80)
```

```
# 显示词云图
word_cloud=word_cloud.fit_words(word_frequence)
plt.imshow(word_cloud)  # 我这里没有中文字体也是乱码
```
